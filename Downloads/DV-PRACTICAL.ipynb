```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataframe
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', periods=5, freq='D'),
    'Stock Price': [100, 101, 102, 103, 102]
})

# Setting the 'Date' column as the index
# data.set_index('Date', inplace=True)

# Plotting the data
plt.figure(figsize=(10, 5))  # Set the figure size
plt.plot(data['Date'], data['Stock Price'], marker='o', linestyle='-', color='b')  # Create line plot
plt.title('Stock Price Over Time')  # Add a title
plt.xlabel('Date')  # Add x-axis label
plt.ylabel('Stock Price')  # Add y-axis label
plt.grid(True)  # Add gridlines
plt.xticks(rotation=45)  # Rotate date labels for better readability
plt.show()  # Display the plot
```

```python
#Scatter-plot

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Seed for reproducibility
np.random.seed(42)

# Generate random data for ages ranging from 20 to 65
ages = np.random.randint(20, 66, size=10)
print(ages)
# Generate income data with some noise to make it more realistic
# Assume income increases with age plus some random noise
base_income = 30000
income = base_income + ages * 750 + np.random.randint(0, 10000, size=10)
print(income)

# Create a DataFrame
data = pd.DataFrame({
    'Age': ages,
    'Income': income
})

# Create the scatter plot
plt.figure(figsize=(10, 5))  # Set the figure size
plt.scatter(data['Age'], data['Income'], color='blue')  # Create scatter points
plt.plot(data['Age'], data['Income'], color='red')


for i, val in enumerate(zip(data["Age"], data["Income"])):
    plt.annotate(str(val), (data["Age"][i], data["Income"][i]))

plt.title('Relationship Between Age and Income')  # Add a title
plt.xlabel('Age')  # Add x-axis label
plt.ylabel('Income')  # Add y-axis label
plt.grid(True)  # Add gridlines
plt.tight_layout()  # Adjust layout to not cut off labels

plt.show()  # Display the plot
```

```python
import matplotlib.pyplot as plt

# Sample data for countries
countries = ['USA', 'China', 'India', 'Germany', 'Brazil']
population = [331, 1441, 1380, 83, 213]  # Population in millions
income = [63000, 10000, 2000, 50000, 15000]  # GDP per capita
life_expectancy = [78, 76, 69, 81, 75]  # Life expectancy in years

# Create bubble chart
plt.figure(figsize=(10, 6))
bubble_size = [le * 100 for le in life_expectancy]  # Adjust size for better visibility

plt.scatter(population, income, s=bubble_size, alpha=0.5, color='blue')
plt.xlabel('Population in Millions')
plt.ylabel('Income in USD')
plt.title('Bubble Chart: Population, Income, and Life Expectancy')

# Annotate each point with country name
for i, country in enumerate(countries):
    plt.annotate(country, (population[i], income[i]))

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Titanic dataset
titanic = sns.load_dataset('titanic')
# Drop rows where 'age' is NaN
titanic = titanic.dropna(subset=['age'])
# Define survivors and non-survivors
survivors = titanic[titanic['survived'] == 1]
non_survivors = titanic[titanic['survived'] == 0]

# Plot histograms
plt.figure(figsize=(10, 6))
plt.hist(survivors['age'], bins=30, alpha=0.5, label='Survived', color='green')
plt.hist(non_survivors['age'], bins=30, alpha=0.5, label='Did Not Survive', color='red')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Age Distribution of Survivors and Non-survivors on the Titanic')
plt.legend()

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the airquality dataset
data = sm.datasets.get_rdataset("airquality").data
# Drop rows with any missing values
data.dropna(inplace=True)
# Calculate the correlation matrix
corr_matrix = data.corr()
# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5, cbar_kws={"shrink": .8})
plt.title('Correlation Heatmap of Air Quality Variables')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset
titanic = sns.load_dataset('titanic')
gender_counts = titanic['sex'].value_counts()
# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, startangle=90, colors=['skyblue', 'lightpink'], autopct='%1.1f%%')

# Draw a circle at the center to turn the pie chart into a donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Optional: Add title
plt.title('Proportion of Male and Female Passengers on the Titanic')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the airquality dataset
data = sm.datasets.get_rdataset("airquality").data
# Drop rows where 'Ozone' is NaN if any
data = data.dropna(subset=['Ozone'])
# Create the strip plot
plt.figure(figsize=(10, 6))
sns.stripplot(y=data['Ozone'], jitter=True, color='blue')  # jitter adds horizontal spread

# Adding labels and title
plt.title('Mean Ozone in Parts Per Billion at Roosevelt Island')
plt.xlabel('Parts Per Billion')  # Simulated categorical axis
plt.ylabel('Ozone')

plt.show()
```
